#### 1) a.

Como as colunas não estão identificadas no dataset *"processed.cleveland.data"* vamos assumir os nomes definidos em *"heart-disease.names"*, secção *"7. Attribute Information"*

```python
import pandas as pd

# From heart-disease.names
columns = ["age", "sex", "cp", "trestbps", "chol", "fbs", "restecg", "thalach", "exang", "oldpeak", "slope", "ca", "thal", "num"]

# Reading the dataset from the local file "processed.cleveland.data" and assigning the column names
df = pd.read_csv("processed.cleveland.data", header=None)
df.columns = columns

df.head()
df.info()
```

NOTA:
`
Pela informação fornecida no documento "heart-disease.names", secção "7. Attribute Information" (linha 109), sabemos que todos os campos são numéricos, no entanto, no bloco anterior verificamos que "ca" e "thal" estão assinalados como "str", daí termos que validar valores corruptos, como strings vazias ou caracteres não-numéricos.
Como essa questão só deve ser analisada posteriomente ( 1) d. ), vou criar uma função apenas para validar se é um valor numérico para ignorar esses valores e posteriormente validarei o que está a falhar.
`

```python
def get_numeric(value):
    try:
        converted = float(value)
        return converted
    except ValueError:
        return None
```

#### 1) b.

```python
discrete_columns = []
for col in df.columns:
    if df[col].dtype == "int64":
        discrete_columns.append(col)
        print(f"{col:>10} -> Discreta")
        continue
    
    is_discrete = True

    for value in df[col]:
        converted = get_numeric(value)
        if converted is not None:
            if converted % 1 > 0:
                is_discrete = False
                break

    if is_discrete:
        discrete_columns.append(col)
        print(f"{col:>10} -> Discreta")
    else:
        print(f"{col:>10} -> Contínua")
```

1) b. **Conclusões:**
* Para analisar se as variáveis são contínuas, sem informação externa ao próprio dataset que explique o significado semântico das variáveis, foram retiradas conclusões apenas com base nos valores presentes. 
* É importante salientar que é conhecido que por exemplo "trestbps" (pressão arterial) é tipicamente um valor numérico contínuo, mas como o objetivo deste trabalho é analisar os dados com base apenas no dataset fornecido e usando as ferramentas de Python para extrair conclusões, serão ignoradas quaisquer deduções semânticas que não sejam representadas nos valores do dataset em estudo, e matemáticamente, valores contínuos estão inseridos no grupos de números reais e os discretos estão no grupo de números inteiros. 
* As variáveis são todas discretas à exceção da variável "oldpeak".
* A variável "oldpeak" é a única que apresenta valores decimais, enquanto as outras variáveis apresentam apenas valores inteiros (apesar de identificados como float64, têm .0 na parte decimal).

```python
nominal_threshold = 10
for col in discrete_columns:
    uniques = df[col].nunique()
    print(f"{col:>10} -> {'Ordinais' if uniques > nominal_threshold else 'Nominais'} : {uniques} valores únicos")
```

1) c. **Conclusões**:
* Como todas as variáveis são numéricas, vamos assumir sendo as variáveis Discretas da alínea anterior como as Categóricas. Daí serão 'age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'slope', 'ca', 'thal' e 'num'.
* Sem um conhecimento semântico das variáveis não é viável atribuir uma natureza nominal ou ordinal apenas com base nos valores do dataset, por isso, apenas para efeitos demonstrativos do exercício defini a regra que se a quantidade de valores unicos for superior a 10 (*nominal_threshold*) então assumimos uma natureza *"Nominal"* em que a ordem não tem influência na informação prestada.
* Depois de analisar as variáveis definidas em *"heart-disease.names"* percebemos que (por acaso) a correspondencia aplica-se em quase todas as variáveis, à exceção da variável *"ca"* (*number of major vessels (0-3) colored by flourosopy*), pois essa representa de facto uma variável Ordinal que pode variar numa escala de 1 a 3, ou seja, a sua ordem tem significado para o sistema.
