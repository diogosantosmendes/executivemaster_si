```python
import pandas as pd

df = pd.read_csv('./data/300110-11-accidentes-bicicleta-csv.csv', sep=';', encoding='utf-8')

display(df.info())
```

```python
nominal_threshold = 10
for col in df.columns:
    if df[col].dtype == "str":
        print(f"{col:>20} -> Categórica Nominal")
        continue
    else:
        is_discrete = True

        for value in df[col]:
            if value % 1 > 0:
                is_discrete = False
                break

    if is_discrete:
        uniques = df[col].nunique()
        print(f"{col:>20} -> Numérica Discreta / Categórica {'Ordinal' if uniques > nominal_threshold else 'Nominal'}")
    else:
        print(f"{col:>20} -> Numérica Contínua")
```

#### 3) d.

Dados recolhidos do pdf:

* 01 Atención en urgencias sin posterior ingreso. - LEVE
* 02 Ingreso inferior o igual a 24 horas - LEVE
* 03 Ingreso superior a 24 horas. - GRAVE
* 04 Fallecido 24 horas - FALLECIDO
* 05 Asistencia sanitaria ambulatoria con posterioridad - LEVE
* 06 Asistencia sanitaria inmediata en centro de salud o mutua - LEVE
* 07 Asistencia sanitaria sólo en el lugar del accidente - LEVE
* 14 Sin asistencia sanitaria
* 77 Se desconoce
* En blanco Sin asistencia sanitaria

Numa primeira análise, a variável parece ser Categórica Nominal, pois surge sob a forma de números (códigos) sem um significado de ordem imediato e sem contexto. No entanto podemos ver que a Câmara Municipal de Madrid já agrupa estes códigos em categorias mais amplas e hierárquicas (LEVE, GRAVE, FALLECIDO), que representam uma progressão na gravidade do desfecho do acidente, desde sair ileso (sem assistência) até ao pior cenário (falecido). Assim, poderia ser realizada uma conversão numa variável Categórica Ordeal.

Vantagens desta abordagem para o seu modelo:
* Reduz o ruído: O modelo não tem de adivinhar a diferença matemática entre "urgência sem internamento" e "assistência ambulatorial", sabendo apenas que ambos são casos "Leves".
* Mantém a ordem matemática: Um valor 3 (Falecido) pesa muito mais do que um 0 (Sem assistência), o que ajuda algoritmos baseados em árvores de decisão (como Random Forest) ou regressões a encontrar padrões mais lógicos. 

Um exemplo de codificação par este processso de conversão seria como descrito no bloco abaixo:

```python
lesividad_map = {
    14: 0,  # Sin asistencia sanitaria
    1: 1,  # LEVE: Atención en urgencias sin posterior ingreso
    2: 1,  # LEVE: Ingreso inferior o igual a 24 horas
    5: 1,  # LEVE: Asistencia sanitaria ambulatoria con posterioridad
    6: 1,  # LEVE: Asistencia sanitaria inmediata en centro de salud
    7: 1,  # LEVE: Asistencia sanitaria sólo en el lugar del accidente
    3: 2,  # GRAVE: Ingreso superior a 24 horas
    4: 3,  # FALLECIDO: Fallecido 24 horas
    77: 0  # SE DESCONOCE: Definimos como 0, pois não temos informação sobre a lesividade
}
# Nans (En blanco) serão tratados como 0 (sem lesividade)
df['lesividad_ordinal'] = df['cod_lesividad'].map(lambda x: lesividad_map.get(x) if x in lesividad_map else 0)
print(df[['cod_lesividad', 'lesividad_ordinal']].head())
```
